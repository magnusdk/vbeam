```python
import matplotlib.pyplot as plt
import numpy as np

from vbeam.postprocess import upsample_by_interpolation

x = np.linspace(0, 1, 6)


def plot_upsampling(n: int):
    upsampled_x = upsample_by_interpolation(x, n)
    plt.scatter(x, np.zeros_like(x) + n, s=30, c="black")
    plt.scatter(upsampled_x, np.zeros_like(upsampled_x) + n, s=15, c="red")
    plt.ylabel("Upsampling factor")
    plt.legend(["Original data", "Upsampled data"])
    plt.ylim([0.5, 7.5])


plot_upsampling(1)
plot_upsampling(2)
plot_upsampling(3)
plot_upsampling(4)
plot_upsampling(5)
plot_upsampling(6)
```
