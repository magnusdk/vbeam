# Window functions
In this notebook we will plot the various window- (apodization-/tapering-) functions implemented in vbeam. The windows are active within the interval [0, 0.5] and are zero everywhere else. To plot the full window (i.e.: from -0.5 to 0.5) we take the absolute value of the argument.

```python
import matplotlib.pyplot as plt
import numpy as np

from vbeam.apodization import (
    Bartlett,
    Hamming,
    Hanning,
    NoWindow,
    Rectangular,
    Tukey25,
    Tukey50,
    Tukey75,
    Tukey80,
    Window,
)

ratio = np.linspace(-1, 1, 200)


def plot_window(window: Window):
    fig, ax = plt.subplots(ncols=2, figsize=(8, 3))
    ax[0].plot(ratio, window(ratio))
    ax[0].set_title("window(ratio)")
    ax[1].plot(ratio, window(np.abs(ratio)))
    ax[1].set_title("window(np.abs(ratio))")
    fig.suptitle(repr(window))
    fig.tight_layout()
```

```python
plot_window(Bartlett())
```

```python
plot_window(Hamming())
```

```python
plot_window(Hanning())
```

```python
plot_window(NoWindow())
```

```python
plot_window(Rectangular())
```

```python
plot_window(Tukey25())
```

```python
plot_window(Tukey50())
```

```python
plot_window(Tukey75())
```

```python
plot_window(Tukey80())
```
